```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_books():
    print("Starting the scraping process...")
    url = "https://books.toscrape.com/catalogue/page-1.html"
    books_data = []

    # Scrape the first 3 pages to get a good dataset
    for page in range(1, 4):
        print(f"Scraping page {page}...")
        current_url = f"https://books.toscrape.com/catalogue/page-{page}.html"
        response = requests.get(current_url)
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Finding all book containers
        books = soup.find_all("article", class_="product_pod")

        for book in books:
            title = book.h3.a["title"]
            price = book.find("p", class_="price_color").text
            rating = book.p["class"][1]  # Extracts 'Three', 'Four', etc.
            availability = book.find("p", class_="instock availability").text.strip()
            
            books_data.append({
                "Title": title,
                "Price": price,
                "Rating": rating,
                "Availability": availability
            })

    # Save to CSV
    df = pd.DataFrame(books_data)
    df.to_csv("CodeAlpha_Books_Dataset.csv", index=False)
    print("Success! Data saved to 'CodeAlpha_Books_Dataset.csv'")

if __name__ == "__main__":
    scrape_books()
```
