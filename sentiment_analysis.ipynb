```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_sentiment():
    # 1. Create a sample dataset of reviews
    data = {
        'Review': [
            "This book was amazing and very helpful!",
            "I hated the plot, it was so boring and slow.",
            "The quality is okay, but it could be better.",
            "Absolutely fantastic experience, highly recommend!",
            "Worst purchase ever, I want my money back.",
            "It was an average read, nothing special.",
            "I love the way the author explains complex topics.",
            "Very disappointed with the delivery and the condition."
        ]
    }
    
    df = pd.DataFrame(data)

    # 2. Function to calculate sentiment
    def get_sentiment(text):
        analysis = TextBlob(text)
        # Polarity ranges from -1 (Negative) to 1 (Positive)
        if analysis.sentiment.polarity > 0:
            return 'Positive'
        elif analysis.sentiment.polarity == 0:
            return 'Neutral'
        else:
            return 'Negative'

    # 3. Apply the analysis
    df['Sentiment'] = df['Review'].apply(get_sentiment)
    
    print("--- Sentiment Analysis Results ---")
    print(df)

    # 4. Visualization
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x='Sentiment', palette='coolwarm', order=['Positive', 'Neutral', 'Negative'])
    plt.title('Sentiment Distribution of Customer Reviews')
    plt.savefig('sentiment_results.png')
    print("\nVisual saved: sentiment_results.png")

if __name__ == "__main__":
    analyze_sentiment()
```
