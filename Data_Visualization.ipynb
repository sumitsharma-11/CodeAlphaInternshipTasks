```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def create_visualizations():
    # Load the cleaned data from our previous steps
    try:
        df = pd.read_csv("CodeAlpha_Books_Dataset.csv")
        df['Price'] = df['Price'].str.replace('£', '').astype(float)
    except Exception as e:
        print(f"Error loading data: {e}")
        return

    # Set the aesthetic style
    sns.set_theme(style="whitegrid")
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Visual 1: Average Price per Rating (Bar Chart)
    # This helps see if higher-rated books are more expensive
    rating_order = ['One', 'Two', 'Three', 'Four', 'Five']
    sns.barplot(ax=axes[0], data=df, x='Rating', y='Price', order=rating_order, palette='magma', ci=None)
    axes[0].set_title('Average Price by Book Rating', fontsize=14)
    axes[0].set_ylabel('Average Price (£)')

    # Visual 2: Price Density (Violin Plot)
    # Shows the distribution and "thickness" of prices across ratings
    sns.violinplot(ax=axes[1], data=df, x='Rating', y='Price', order=rating_order, palette='viridis')
    axes[1].set_title('Price Distribution Density per Rating', fontsize=14)
    axes[1].set_ylabel('Price (£)')

    plt.tight_layout()
    plt.savefig('Task3_Data_Story.png')
    print("Success! Professional dashboard saved as 'Task3_Data_Story.png'")

if __name__ == "__main__":
    create_visualizations()
```
